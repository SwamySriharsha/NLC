```python
import PyPDF2
import re
from datetime import datetime

scrapdata = open('webscrapdatamod.pdf', 'rb')

read_pdf = PyPDF2.PdfFileReader(scrapdata)

nop = read_pdf.getNumPages()

pdfWriter = PyPDF2.PdfFileWriter()

print(nop)
```

```python


def modify(testtime):
    testtime
    tf2 = re.findall('\\b\d{1,2}\:\d{1,2}(?:am|PM|pm|AM)',testtime) #H:MM
    #print(tf2)
    for i in tf2:
        try:
            objDate = datetime.strptime(i, '%H:%M%p')
            fstr = datetime.strftime(objDate,'%R')
            testtime = re.sub('\\b\d{1,2}\:\d{1,2}(?:am|PM|pm|AM)',fstr,testtime)
        except:
            print("Invalid time is {0}".format(i))
            
    tf3 = re.findall('\\b\d{1,2}\.\d{1,2}(?:PM|am|AM|pm)',testtime) #H.MM 
    #print(tf3)
    for i in tf3:
        try:
            objDate = datetime.strptime(i, '%H.%M%p')
            fstr = datetime.strftime(objDate,'%R')
            testtime = re.sub('\\b\d{1,2}\.\d{1,2}(?:PM|am|AM|pm)',fstr,testtime)
        except:
            print("Invalid time is {0}".format(i))
    print(testtime)       

for p in range(nop):
    page = read_pdf.getPage(p)
    testtime = page.extractText()
    modify(testtime)
    
    
    
```

```python
def modifynum(mfre1):
    for i in mf1:
        cpn(i,mfre1)
        
def cpn(mobile,mfre):
    global testmob
    valid_prefixes = '[+91|91|(+91)|( +91 )]'
    result = re.match('^'+valid_prefixes+'*([0-9]{10})$', mobile)
    if result:
        nn = '+91' + result.group(1)
        testmob = re.sub(mfre,nn,testmob)
        print(testmob)
    else:
        print('Invalid mobile num {0}'.format(mobile))
        

page = read_pdf.getPage(4)
testmob = page.extractText()
mfre1 = '\(? ?\+ ?91 ?\)? ?\d{9,10}'
mf1 = re.findall(mfre1,testmob)
print(mf1)
modifynum(mfre1)
    
```
